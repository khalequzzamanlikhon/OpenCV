```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
print(cv2.__version__)
```

```python
img=cv2.imread("images/cards.jpg")
width,height=250,350 # width and heigh after perspective transfromation
pts1 = np.float32([[111,219],[287,188],[154,482],[352,440]])
pts2 = np.float32([[0,0],[width,0],[0,height],[width,height]])

matrix=cv2.getPerspectiveTransform(pts1,pts2)
imgOutput=cv2.warpPerspective(img,matrix,(width,height))
```

- pts1: Four source points in the original image representing the coordinates of the corners of a region to be transformed.
- pts2: Four destination points representing where the corresponding source points should be mapped to in the output image.
- cv2.getPerspectiveTransform(pts1, pts2): Computes the perspective transformation matrix using the source and destination points.

```python
#display image
cv2.imshow("Original image",img)
cv2.imshow("output",imgOutput)
cv2.waitKey(0)
cv2.destroyAllWindows()


# plot the image

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original iamge")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(imgOutput)
plt.title("Output image")
plt.axis("off")
```
