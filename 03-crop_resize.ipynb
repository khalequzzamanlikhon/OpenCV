```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
print(cv2.__version__)
```

```python
img=cv2.imread("images/mypic1.jpg")
cv2.imshow("original Image",img)
# resize
resized_img=cv2.resize(img,(600,400))
cv2.imshow("resized image",resized_img)
#cropping
cropped_img=img[80:400,80:600]
cv2.imshow("cropped image",cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(f"shape of original iamge{img.shape}")
print(f"shape after resize {resized_img.shape}")
print(f"shape after cropping {cropped_img.shape}")

# showing the images

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("original image")
plt.axis("off")


plt.subplot(1,3,2)
plt.imshow(resized_img)
plt.title("rsized image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cropped_img)
plt.title("cropped image")
plt.axis("off")

plt.show()




```
